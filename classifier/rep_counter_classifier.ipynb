```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
# Planks
import time

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    if angle > 180.0:
        angle = 360 - angle
    return angle

cap = cv2.VideoCapture(0)

cumulative_time = 0
start_time = None
timer_running = False

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = pose.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        try:
            landmarks = results.pose_landmarks.landmark

            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                        landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                   landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                     landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]

            body_angle = calculate_angle(shoulder, hip, ankle)

            cv2.putText(image, f'Angle: {int(body_angle)}',
                        tuple(np.multiply(hip, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)

            if 160 < body_angle < 195:
                if not timer_running:
                    start_time = time.time()
                    timer_running = True
                elapsed_time = time.time() - start_time
            else:
                if timer_running:
                    # Add the last hold duration to cumulative time
                    cumulative_time += time.time() - start_time
                timer_running = False
                elapsed_time = 0

            total_time = cumulative_time + (elapsed_time if timer_running else 0)

            cv2.putText(image, f'Time: {total_time:.1f}s', (10, 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0) if timer_running else (0, 0, 255), 3, cv2.LINE_AA)

        except Exception as e:
            timer_running = False
            elapsed_time = 0
            total_time = cumulative_time
            # Optionally print or log e

        if results.pose_landmarks:
            mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        cv2.imshow('Plank Timer', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python


mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# Helper function: calculate angle between 3 points (if needed)
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    if angle > 180.0:
        angle = 360 - angle
    return angle

cap = cv2.VideoCapture(0)

counter = 0
stage = "in"  # initial stage - legs together

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = pose.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        try:
            landmarks = results.pose_landmarks.landmark

            # Get relevant landmarks
            left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                        landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                         landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
            left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                          landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]

            # Calculate horizontal distances from center (hip midpoint)
            center_x = (left_hip[0] + right_hip[0]) / 2

            # Distances from center for knees to detect step out
            left_knee_dist = left_knee[0] - center_x
            right_knee_dist = right_knee[0] - center_x

            # Threshold for detecting side lunge (tweak if needed)
            step_out_threshold = 0.15  # approx 15% of frame width

            # Visualize distances
            cv2.putText(image, f"L Dist: {left_knee_dist:.2f}", (10,30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 2)
            cv2.putText(image, f"R Dist: {right_knee_dist:.2f}", (10,60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 2)

            # Logic to detect lunges
            if stage == "in":
                if right_knee_dist > step_out_threshold:
                    stage = "out_right"
                elif left_knee_dist < -step_out_threshold:
                    stage = "out_left"

            elif stage == "out_right":
                if right_knee_dist <= step_out_threshold / 2:
                    stage = "in"
                    counter += 1
                    print(f" Side Lunge Right #{counter}")

            elif stage == "out_left":
                if left_knee_dist >= -step_out_threshold / 2:
                    stage = "in"
                    counter += 1
                    print(f" Side Lunge Left #{counter}")

        except:
            pass

        # Draw info box
        cv2.rectangle(image, (0,0), (300, 90), (245,117,16), -1)

        cv2.putText(image, 'SIDE LUNGES', (10,20), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,0), 2)
        cv2.putText(image, 'REPS', (15,55), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,0), 2)
        cv2.putText(image, str(counter), (100, 55), 
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2)
        cv2.putText(image, 'STAGE', (170, 55), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,0), 2)
        cv2.putText(image, stage, (240, 55), 
                    cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 2)

        # Draw pose landmarks
        if results.pose_landmarks:
            mp_drawing.draw_landmarks(
                image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
            )

        cv2.imshow('Side Lunge Counter', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
